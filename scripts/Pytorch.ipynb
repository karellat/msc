```python
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from deep_mri.dataset.dataset_3d import _get_label_str
from deep_mri.dataset.dataset import get_train_valid_files
from deep_mri.dataset.dataset import numpy_to_nibabel
import nibabel as nib
import numpy as np
from torchviz import make_dot
from classifier import LinearClassifierAlexNet
from models import NetWork
from LinearModel import alex_net_complete
import torch.optim as optim


class ToTensor(object):
    """Convert ndarrays in sample to Tensors."""

    def __call__(self, sample):
        image, label = sample['image'], sample['label']

        # swap color axis because
        # numpy image: H x W x C
        # torch image: C X H X W
        image = image.transpose((3, 0, 1, 2))
        return {'image': torch.from_numpy(image),
                'label': torch.from_numpy(label)}

def decode_img(path, normalize): 
    img = nib.load(path).get_fdata().squeeze()
    img[np.isnan(img)] = 0.0
    if normalize: 
        img = (img - img.min()) / (img.max() - img.min() + 1e-6)
    return np.array(np.expand_dims(img, axis=-1), dtype=np.float32)
    
    
class dataset_3d(Dataset):
    def __init__(self,
                 scan_files, 
                 normalize=True, 
                 transform=None):
        
        self.scan_files  = scan_files
        self.transform = transform
        self.normalize = normalize

        return 
    def __len__(self):
        return len(self.scan_files)
    
    def __getitem__(self, idx):
        img_path = self.scan_files[idx]
        label = np.array(
            np.argmax(_get_label_str(img_path))
        )
        img = decode_img(img_path, self.normalize)
        
        sample = {
            'image' : img,
            'label' : label
        }
        
        if self.transform:
            sample = self.transform(sample)
        
        return sample
```

```python
import time
import copy
def train_model(model, criterion, optimizer, num_epochs=25):
    since = time.time()

    best_model_wts = copy.deepcopy(model.state_dict())
    best_acc = 0.0

    for epoch in range(num_epochs):
        print('Epoch {}/{}'.format(epoch, num_epochs - 1))
        print('-' * 10)

        # Each epoch has a training and validation phase
        for phase in ['train', 'val']:
            if phase == 'train':
                model.train()  # Set model to training mode
            else:
                model.eval()   # Set model to evaluate mode

            running_loss = 0.0
            running_corrects = 0

            # Iterate over data.
            for data in dataloaders[phase]:
                inputs, labels = data['image'], data['label']
                inputs = inputs.to(device)
                labels = labels.to(device)

                # zero the parameter gradients
                optimizer.zero_grad()

                # forward
                # track history if only in train
                with torch.set_grad_enabled(phase == 'train'):
                    outputs = model(inputs, None)
                    _, preds = torch.max(outputs, 1)
                    loss = criterion(outputs, labels)
                    # backward + optimize only if in training phase
                    if phase == 'train':
                        loss.backward()
                        optimizer.step()

                # statistics
                running_loss += loss.item() * inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)
 

            epoch_loss = running_loss / dataset_sizes[phase]
            epoch_acc = running_corrects.double() / dataset_sizes[phase]

            print('{} Loss: {:.4f} Acc: {:.4f}'.format(
                phase, epoch_loss, epoch_acc))

            # deep copy the model
            if phase == 'val' and epoch_acc > best_acc:
                best_acc = epoch_acc
                best_model_wts = copy.deepcopy(model.state_dict())

        print()

    time_elapsed = time.time() - since
    print('Training complete in {:.0f}m {:.0f}s'.format(
        time_elapsed // 60, time_elapsed % 60))
    print('Best val Acc: {:4f}'.format(best_acc))

    # load best model weights
    model.load_state_dict(best_model_wts)

    return model
```
