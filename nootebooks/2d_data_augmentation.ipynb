```python
from deep_mri.model_zoo import model_factory

model = model_factory("2d_transfer", num_outputs=3, feature_extractor_url="https://tfhub.dev/tensorflow/efficientnet/b3/feature-vector/1")
```

```python
log_dir = 'test_dir'
from deep_mri.train.callbacks import ConfusionMatrixCallback, DummyPredictorCallback, SaveConfigCallback
from deep_mri.dataset import CLASS_NAMES
callbacks = [
    tf.keras.callbacks.TensorBoard(log_dir=log_dir,
                                           histogram_freq=0,
                                           write_graph=False),
    ConfusionMatrixCallback(valid_ds, CLASS_NAMES, log_dir), 
    SaveConfigCallback(log_dir, {'example' :'neco'}),
    DummyPredictorCallback(train_ds, valid_ds)
]
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(),
                      loss=tf.keras.losses.CategoricalCrossentropy(),
                      metrics=[tf.keras.metrics.CategoricalAccuracy(), 
            tf.keras.metrics.AUC()])

```

```python
BATCH_SIZE = 32
EPOCHS = 3
model.fit(train_ds.batch(BATCH_SIZE), 
          validation_data=valid_ds.batch(BATCH_SIZE),
          epochs=EPOCHS,
          verbose=1,
          workers=16,
          use_multiprocessing=True,
          callbacks=callbacks)
```
